```python
import cv2
import numpy as np
```

```python
kernel = np.ones((3,3))/9.0
```

```python
kernel
```

```python
img = cv2.imread("data/src/Lena.jpg",0)
img_ke1 = cv2.filter2D(img,-1,kernel)
```

```python
cv2.imshow("img",img_ke1)
cv2.imshow("src",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
kernel2 = np.zeros((3,3))
kernel2[0,0] = 1
kernel2[1,0] = 2
kernel2[2,0] = 3
kernel2[0,2] = -1
kernel2[1,2] = -2
kernel2[2,2] = -3
```

```python
kernel2
```

```python
img_ke2 = cv2.filter2D(img,-1,kernel2)
```

```python
cv2.imshow("img",img_ke2)
cv2.imshow("src",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
